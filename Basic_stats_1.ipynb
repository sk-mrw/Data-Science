## describe()
### This method of pandas library used for getting an statistic overview of the data stored in dataframe.
##### Count ---> It returns the number of data point available in that particular column
##### mean ----> It returns average calculated value of that column
##### std ----> It returns standard deviation i.e., avarage deviation from mean 
##### min ----> It returns minimum value in the entire column
##### 25% ----> It returns first quartile value 25%th value of that column
##### 50% -----> It returns median value of the column
##### 75% -----> It returns third qurtile value 75%th value of that column
##### max -----> It returns Max value in the entire column

## Boxplots

#### Outlier
- Outliers are the exceptional values which is either too high or too which could affect entire dataset.
- Detecting outliers is very important for model acuracy
- Boxplot is the best way to check outliers.
- The dots showing above upper extreme or the dots showing below lower extreme are considered as outliers.
- In our dataset's Brand column Babaji Brand has 2 outliers, Gear has 1 outlier and Jeer has 3 outliers.

```python
import seaborn as sns
sns.boxplot(x=df["Brand"],y=df["Volume"],data=df)
```

##### Z-score normalization
- In our data the columns would have some units so, it is not appropriate to use it directly for training machine learning model.
- So, we normalize the data using the difference from mean divided by std removes the units from the given X variable. so that our variables becomes free units.

##### One hot encoding: 
- It is a data preprocessing technique used for converting categorical data into boolean to make the data suitable for machine learning models
- In this process Values of one column are converted to columns and in those columns 1 or 0 is filled based on the original column.
- get_dummies() of pandas is used for the similar practise.
- Additionally, by defalult it drops the first column to avoid multicolliniearity (one column can be perfectly predicted by the others—which messes with how the model estimates coefficients)



```python
df[df.columns[[0,1,2,3,5,6,7]]].describe()
```

```python
new=pd.get_dummies(df[df.columns[[0,1,2,3,5,6,7]]],drop_first="True")
```
