```python
# Now we are going to update df for the unknown with mode
df=df.apply(lambda x: x.str.strip() if x.dtype=='object' else x) # Trim spaces
for col in ['workclass', 'occupation', 'native_country']:
    df[col] = df[col].replace('?', df[col].mode()[0])
```

#### Scaling for numerical columns:
##### Standard Scaling:
- When to use: Preferred when the feature follows a Normal (Gaussian) distribution.
- It is essential for algorithms like SVM, Logistic Regression, and PCA that assume centered data.
- Why: It centers data around a mean of 0 with a standard deviation of 1.
##### Min-Max Scaling:
- When to use: Preferred when you need the data within a specific bounded range (usually 0 to 1) and when the distribution is not Gaussian.
- Ideal for KNN or Neural Networks.Why: It preserves the relative distances between values but is sensitive to extreme outliers.



```python
# Task 2: Encoding
# One-Hot (< 5 categories): sex, income
# Label Encoding (>= 5 categories): workclass, education, marital_status, occupation, relationship, race, native_country

cat_counts = df.select_dtypes(include=['object']).nunique()
one_hot_cols = cat_counts[cat_counts < 5].index.tolist()
label_encode_cols = cat_counts[cat_counts >= 5].index.tolist()

print(f"One-Hot columns: {one_hot_cols}")
print(f"Label Encode columns: {label_encode_cols}")

# Apply One-Hot
df_encoded = pd.get_dummies(df, columns=one_hot_cols,dtype=int)

# Apply Label Encoding
le = LabelEncoder()
for col in label_encode_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])

print(df_encoded.head())
```

We use df_encoded to maintain data integrity. Keeping the original dataset (df) separate allows us to perform comparative analysis and ensures that we can always refer back to the original categorical labels if needed for interpretability, while providing the model with the numeric format it requires.

#### Encoding Techniques
##### Following the instruction to use One-Hot Encoding for categories < 5 and Label Encoding for categories >= 5:
- One-Hot Encoded: sex, income.
- Label Encoded: workclass, education, marital_status, occupation, relationship, race, native_country.
#### Pros and Cons:
#### One-Hot Encoding:
##### Pros: No artificial ordering is introduced. Works well with linear models.
##### Cons: Can lead to the "curse of dimensionality" if a feature has many categories.

#### Label Encoding:
##### Pros: Memory efficient; keeps the number of columns the same.
##### Cons: Introduces a mathematical order ($0 < 1 < 2$) that doesn't exist in reality (e.g., "Doctor" vs "Engineer"), which can confuse some models.

#### Task 3: Feature Engineering
- New Feature 1: `net_capital` Rationale: capital_gain and capital_loss are related. Creating a net value represents the actual financial growth/loss in a single feature, reducing complexity.
- New Feature 2: `age_group` Rationale: Income patterns often change in stages (e.g., entry-level, peak career, retirement). Binning age into categories like 'Young', 'Middle-aged', 'Senior' captures these life stages better than a linear age.
##### Log Transformation:
- Feature: capital_gain (Skewness: 11.95). Rationale: The data is highly right-skewed. Log transformation ($log(1+x)$) compresses the range and makes the distribution more "normal," which helps models converge faster.




```python
corr_matrix = df_encoded.corr()
print("Correlation with income_>50K:")
print(corr_matrix['income_>50K'].sort_values(ascending=False))
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap='coolwarm')
plt.title("Correlation Matrix")
```

```python
print(df[num_col].skew())
```

```python
# Feature Engineering
df_fe = df.copy()

# New Feature: Net Capital
df_fe['net_capital'] = df_fe['capital_gain'] - df_fe['capital_loss']

# New Feature: Age Bin
df_fe['age_group'] = pd.cut(df_fe['age'], bins=[0, 25, 45, 65, 100], labels=['Young', 'Middle-aged', 'Senior', 'Old'])

# Log Transformation on skewed feature 'capital_gain'
df_fe['log_capital_gain'] = np.log1p(df_fe['capital_gain'])

print("Task 3 Features Created:")
print(df_fe[['net_capital', 'age_group', 'log_capital_gain']].head())
print("\n",df_fe[['net_capital', 'age_group', 'log_capital_gain']].nunique())
```

```python
plt.figure(figsize=(15,10))
for i,col in enumerate(num_col):
    plt.subplot(3,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(f"Distribution of {col}")

plt.tight_layout()
```

```python
# We need to use numeric data for Isolation Forest
# Let's use the df_encoded from before but with FE features added/encoded

# Preparing data for Isolation Forest
# 1. New Feature: Net Capital (Rationale: Combines gain/loss into a single economic indicator)
df_encoded['net_capital'] = df['capital_gain'] - df['capital_loss']

# 2. New Feature: Age Group (Rationale: Captures life-stage earning potential)
df_encoded['age_group_label'] = pd.cut(df['age'], bins=[0, 25, 45, 65, 100], labels=[0, 1, 2, 3]).astype(int)

# 3. Log Transformation (Rationale: Reduces skewness of capital_gain for better model performance)
df_encoded['log_capital_gain'] = np.log1p(df['capital_gain'])

iso_data = df_encoded.select_dtypes(include=[np.number])
iso = IsolationForest(contamination=0.05, random_state=42)
outliers = iso.fit_predict(iso_data)

# Remove outliers
df_clean = df_encoded[outliers == 1]
print(f"\nOriginal rows: {len(df_encoded)}")
print(f"Rows after removing outliers: {len(df_clean)}")
print(f"Outliers removed: {len(df_encoded) - len(df_clean)}")

plt.figure(figsize=(12, 8))
sns.heatmap(df_clean.corr(), cmap='coolwarm')
plt.title("Correlation Matrix after Cleaning")
```

```python
for i,col in enumerate(['capital_gain','log_capital_gain']):
    plt.subplot(1,2,i+1)
    sns.histplot(df_encoded[col],kde=True)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
```
