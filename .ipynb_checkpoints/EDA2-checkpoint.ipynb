In the above cell, I have observed there is an extra space in each of the categorical features and ther are also '?' in many features.

We will replace the '?' value with the mode of that feature

```python
for col in cat_col:
    unique_counts = df[col].value_counts()
    if '?' in unique_counts:
        print(f"{col} contains {unique_counts['?']} '?' values out of Total {df[col].count()}")
    
```

```python
# Now we are going to update df for the unknown with mode
df=df.apply(lambda x: x.str.strip() if x.dtype=='object' else x) # Trim spaces
for col in ['workclass', 'occupation', 'native_country']:
    df[col] = df[col].replace('?', df[col].mode()[0])
```

```python
for i in cat_col:
    print(f"values in {i} column\n{df[i].unique()}\n")
```

```python
# Scaling Numeric value
scaler_std = StandardScaler()
df_std = df.copy()
df_std[num_col] = scaler_std.fit_transform(df[num_col])

scaler_minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[df_num_col] = scaler_minmax.fit_transform(df[num_col])

```

```python
# Task 2: Encoding
# One-Hot (< 5 categories): sex, income
# Label Encoding (>= 5 categories): workclass, education, marital_status, occupation, relationship, race, native_country

cat_counts = df.select_dtypes(include=['object']).nunique()
one_hot_cols = cat_counts[cat_counts < 5].index.tolist()
label_encode_cols = cat_counts[cat_counts >= 5].index.tolist()

print(f"One-Hot columns: {one_hot_cols}")
print(f"Label Encode columns: {label_encode_cols}")

# Apply One-Hot
df_encoded = pd.get_dummies(df, columns=one_hot_cols,dtype=int)

# Apply Label Encoding
le = LabelEncoder()
for col in label_encode_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])

print("\nTask 1 & 2 Completed.")

print(df_encoded.head())
```

We use df_encoded to maintain data integrity. Keeping the original dataset (df) separate allows us to perform comparative analysis and ensures that we can always refer back to the original categorical labels if needed for interpretability, while providing the model with the numeric format it requires.

```python
# Task 3: Feature Engineering
df_fe = df.copy()

# 1. New Feature: Net Capital
df_fe['net_capital'] = df_fe['capital_gain'] - df_fe['capital_loss']

# 2. New Feature: Age Bin
df_fe['age_group'] = pd.cut(df_fe['age'], bins=[0, 25, 45, 65, 100], labels=['Young', 'Middle-aged', 'Senior', 'Old'])

# 3. Log Transformation on skewed feature 'capital_gain'
df_fe['log_capital_gain'] = np.log1p(df_fe['capital_gain'])

print("Task 3 Features Created:")
print(df_fe[['net_capital', 'age_group', 'log_capital_gain']].head())

# Task 4: Feature Selection - Isolation Forest
# We need to use numeric data for Isolation Forest
# Let's use the df_encoded from before but with FE features added/encoded

# Preparing data for Isolation Forest
iso_data = df_encoded.select_dtypes(include=[np.number])
iso = IsolationForest(contamination=0.05, random_state=42)
outliers = iso.fit_predict(iso_data)

# Remove outliers
df_no_outliers = df_encoded[outliers == 1]

print(f"\nOriginal rows: {len(df_encoded)}")
print(f"Rows after removing outliers: {len(df_no_outliers)}")
print(f"Outliers removed: {len(df_encoded) - len(df_no_outliers)}")
```
