```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
# Encode categorical variables
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])

# Visualize class distribution
sns.countplot(x='class', data=df)
plt.title('Class Distribution (0 = Edible, 1 = Poisonous)')
plt.show()
X=df.drop(['Unnamed: 0','class'],axis=1)
y = df['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
df.columns
```

```python
# Since there are 25 features, a single pairplot is too big.
# We use a loop to visualize subsets of features against numerical ones.

# Example loop function to generate multiple pairplots for different feature groups
feature_groups = [
    ['stalk_height', 'cap_diameter', 'odor', 'class'],
    ['cap_shape', 'cap_surface', 'cap_color', 'class'],
    ['gill_attachment', 'gill_spacing', 'gill_size', 'class']
]

for i, group in enumerate(feature_groups):
    sns.pairplot(df[group], hue='class', palette='husl')
    plt.suptitle(f'Feature Group {i+1} Correlation Pairplot', y=1.02)
    print(plt.show())
    plt.close() # Close to save memory
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix


# Implement a basic SVM classifier (using RBF kernel by default)
svm_model = SVC(kernel='rbf', C=1.0)
svm_model.fit(X_train, y_train)

# Evaluate on testing data
y_pred = svm_model.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(classification_report(y_test, y_pred, zero_division=np.nan))

# Visualize Results with a Confusion Matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Experiment with different kernels and C values
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'poly', 'rbf'],
    'gamma': ['scale', 'auto']
}

grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2,cv=3,n_jobs=-1)
grid.fit(X_train, y_train)

print(f"Best Parameters: {grid.best_params_}")
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(classification_report(y_test, y_pred, zero_division=np.nan))

# Visualize Results with a Confusion Matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

### Strengths for Mushroom Data: SVM is highly effective in high-dimensional spaces (many categorical features) and is robust against overfitting.
### Kernels: While a Linear kernel is fast, the RBF or Polynomial kernels often perform better if the boundary between poisonous and edible mushrooms is complex.
### Practical Implications: In real-world classification, SVM's ability to maximize the margin (the distance between classes) makes it reliable for safety-critical tasks like identifying poisonous substances.
